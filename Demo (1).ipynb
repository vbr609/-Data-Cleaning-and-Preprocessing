```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Example DataFrame with missing values
data = {'A': [1, 2, None, 4], 'B': [5, None, 7, 8], 'C': [None, 2, 3, 4]}
df = pd.DataFrame(data)

# Define the imputer
impute = SimpleImputer(strategy='mean')  # You can choose other strategies too

# Apply the imputation
for i in df.select_dtypes(include="number").columns:
    df[i] = impute.fit_transform(df[[i]])

print(df)
```

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```
